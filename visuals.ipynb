```python
import numpy as np
import pandas as pd
import rho_plus as rp
import seaborn as sns


theme, cs = rp.mpl_setup(False)
```

```python
data = [
    [79, 11, 2, 4, 2, 1, 1],
    [8, 69, 15, 5, 1, 1, 2],
    [2, 13, 78, 4, 0, 1, 2],
    [2, 4, 4, 84, 1, 4, 1],
    [4, 2, 1, 2, 90, 2, 1],
    [1, 2, 2, 5, 1, 87, 2],
    [1, 1, 3, 2, 0, 0, 93]
]

systems = ['Triclinic', 'Monoclinic', 'Orthorhombic', 'Tetragonal', 'Trigonal', 'Hexagonal', 'Cubic']

df = pd.DataFrame(data, index=systems, columns=systems)
ax = sns.heatmap(df / 100, cmap='rho_heatmap', annot=True, fmt='.0%', cbar=False, square=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', color='k')
ax.set_yticklabels(ax.get_yticklabels(), color='k')
ax.set_ylabel('True System')
ax.set_xlabel('Predicted System')
rp.ylabel_top()
ax.set_title('Confusion Matrix')
```
